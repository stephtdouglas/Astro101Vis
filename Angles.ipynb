```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Wedge, Arc
```

```python
circ1 = Circle((0,0),radius=3,edgecolor="k",facecolor="none")
wedge_degree = Wedge((0,0),3,-0.5,0.5,edgecolor="C0",facecolor="none",linewidth=0.5)

plt.figure(figsize=(5,5))
ax = plt.subplot(111)
ax.add_patch(circ1)
ax.add_patch(wedge_degree)
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
# ax.set_xlim(0,3.2)
# ax.set_ylim(-1.6,1.6)
plt.axis("off")
```

```python
circ1 = Circle((0,0),radius=3,edgecolor="k",facecolor="none",zorder=20)
wedge_degree = Wedge((0,0),3,-5,5,edgecolor="C0",facecolor="w")
# wedge_degree_zoom = Wedge((2.45,0),4,-24,24,edgecolor="w",facecolor="w",zorder=10)
wedge_degree_zoom2 = Wedge((2.45,0),4.5,-25,25,edgecolor="C0",facecolor="none",zorder=-3)

plt.figure(figsize=(10,5))
ax = plt.subplot(111)
# ax.add_patch(wedge_degree_zoom)
ax.add_patch(wedge_degree_zoom2)

start = -25
step = 50/6
for i in range(6):
  stop = start+step
  ten_arcmin = Wedge((2.45,0),4.5,start,stop,edgecolor="C1",facecolor="none",zorder=-5)
  ax.add_patch(ten_arcmin)
  start += step
start = -25
step = 50/6/10
for i in range(10):
  stop = start+step
  arcmin = Wedge((2.45,0),4.5,start,stop,edgecolor="C1",facecolor="none",zorder=-5)
  ax.add_patch(arcmin)
  start += step

ax.add_patch(circ1)
ax.add_patch(wedge_degree)

# That gap at 0 degrees does not appear to be caused by the Arc, it might be the white wedge cover
one_deg_arc = Arc((0,0),4.5,4.5,0,-5,5,color="C0",linestyle="-")
ax.add_patch(one_deg_arc)

one_deg_arc2 = Arc((2.45,0),3.5,3.5,0,-25,25,color="C0",linestyle="-")
ax.add_patch(one_deg_arc2)

arcmin_zoom = Wedge((6.5,-1.9),4.5,-25,25,edgecolor="C1",facecolor="none",zorder=-3)
ax.add_patch(arcmin_zoom)


start = -25
step = 50/6
for i in range(6):
  stop = start+step
  ten_arcsec = Wedge((6.5,-1.9),4.5,start,stop,edgecolor="C2",facecolor="none",zorder=-5)
  ax.add_patch(ten_arcsec)
  start += step
start = -25
step = 50/6/10
for i in range(10):
  stop = start+step
  arcsec = Wedge((6.5,-1.9),4.5,start,stop,edgecolor="C2",facecolor="none",zorder=-5)
  ax.add_patch(arcsec)
  start += step

arcmin_arc = Arc((6.5,-1.9),3.5,3.5,0,-25,25,color="C1",linestyle="-")
ax.add_patch(arcmin_arc)

ax.set_xlim(-3,13)
ax.set_ylim(-4,4)
# ax.set_xlim(0,3.2)
# ax.set_ylim(-1.6,1.6)
plt.axis("off")
```
