# Scatter plots and trendlines

```python
x1 = np.linspace(1e-2,10,100)
y1 = x1

x2 = x1
y2 = x2**2

rng = np.random.default_rng(seed=42)
x3 = rng.choice(x1,15,replace=False)
y3 = x3 + rng.normal(size=15)
```

```python
fig = plt.figure(figsize=(9,4.5))
ax = fig.add_subplot(111)
plt.plot(x3,y3,'o',ms=10)
ax.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
fig.savefig("linear_scatter.png",dpi=600,bbox_inches="tight")
plt.plot(x1,y1,'-',lw=2)
fig.savefig("linear_scatter_trend.png",dpi=600,bbox_inches="tight")
```

```python
fig = plt.figure(figsize=(9,4.5))
ax = fig.add_subplot(111)
ax.grid(True)
plt.xlabel("X")
plt.ylabel("Y")
plt.ylim(-0.5,10.5)
plt.plot(x1,y1,'--',lw=2,color="C1")
fig.savefig("linear_trend.png",dpi=600,bbox_inches="tight")
plt.plot(x1,np.log(x1)+3,'-',lw=2,color="C2")
fig.savefig("linear_functions.png",dpi=600,bbox_inches="tight")
```
